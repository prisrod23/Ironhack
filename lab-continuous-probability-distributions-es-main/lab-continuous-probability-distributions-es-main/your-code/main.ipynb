```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

def generate_uniform_random(bottom, ceiling, count):
    return uniform.rvs(loc=bottom, scale=ceiling-bottom, size=count) #.tolist()

# Generar números aleatorios con diferentes parámetros
random1 = generate_uniform_random(10, 15, 100)

random2 = generate_uniform_random(10, 60, 1000)

# Crear histogramas
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(random1, bins=100, edgecolor='black')
plt.title('Distribución Uniforme (10-15, 100)')
plt.xlabel('Valor')
plt.ylabel('Recuento')

plt.subplot(1, 2, 2)
plt.hist(random2, bins=100, edgecolor='black')
plt.title('Distribución Uniforme (10-60, 1000)')
plt.xlabel('Valor')
plt.ylabel('Recuento')

plt.tight_layout()
plt.show()
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
def generate_exponential_data(mean, size):
    """
    Genera una lista de números distribuidos exponencialmente.
    
    Args:
        mean (float): La media de la distribución exponencial.
        size (int): El tamaño de la lista a generar.
    
    Returns:
        list: Una lista de números distribuidos exponencialmente.
    """
    return np.random.exponential(scale=mean, size=size)

# Generar dos secuencias de números exponencialmente distribuidos
data_10 = generate_exponential_data(1, 100)
data_100 = generate_exponential_data(10, 100)

# Trazar los histogramas
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

ax1.hist(data_10, bins=100, edgecolor='black', color='yellow')
ax1.set_title('Distribución Exponencial (n=10)')
ax1.set_xlabel('Valor')
ax1.set_ylabel('Frecuencia')

ax2.hist(data_100, bins=100, edgecolor='black', color='green')
ax2.set_title('Distribución Exponencial (n=100)')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuencia')


plt.show()
```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia. 

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
import math

# Parámetro de la distribución exponencial (λ = 1/10)
lambda_value = 1 / 10

# Valor para el que queremos calcular la probabilidad (x = 15)
x_value = 15

# Calculamos la probabilidad
probability = 1 - math.exp(-lambda_value * x_value)
```

¿Cuál es la probabilidad de que el cliente pase más de 15 minutos?

```python
print(f"La probabilidad de que un cliente pase menos de 15 minutos en el banco es: {probability:.4f}")
```
